```python
# Make a copy of the original DataFrame
df_copy = df.copy()
columns_list = [
    "Ad set budget", 
    "Attribution setting", 
    "Results", 
    "Result indicator", 
    "Reach", 
    "Impressions", 
    "Cost per results", 
    "Amount spent (USD)", 
    "Reporting starts", 
    "Reporting ends", 
    "campaign_type"
]

df_copy = df_copy.drop(columns=columns_list)

# Debugging: Check what the regex catches 
# worked
# df_copy['check'] = df_copy['content'].str.extract(r'(?i)\b(post)\b[\s;:]*', expand=False)

df_copy

df_copy['category'] = df_copy['content'].str.extract(r'(?i)\b(post)\b[\s;:]*', expand=False)

df_copy['category'] = df_copy['category'].fillna(df_copy['content'])


# Count occurrences of each category
category_counts = df_copy['category'].value_counts()

print(category_counts)

```

```python
cleaned_df = df.drop(['Reporting ends','Attribution setting', 'Reporting starts','Reporting ends', 'content', 'Ad set budget'], axis=1)
```

```python
cleaned_df = cleaned_df.rename(columns={'campaign_type': 'campaign type'})
cleaned_df.head()
```

```python
cleaned_df1 = cleaned_df.drop(['Cost per results'], axis=1)
cleaned_df1.head()
```

```python
cleaned_df1['cost_per_result'] = ( cleaned_df1['Amount spent (USD)']/cleaned_df1['Results'])
cleaned_df1 = cleaned_df1.drop(['Cost/Result'], axis=1)

cleaned_df1.head()
```

```python
cleaned_df1.rename(columns= {'Results': 'results',	'Result indicator': 'result_indicator',	'Reach': 'reach',	'Impressions': 'impressions',
                             'Amount spent (USD)': 'amount_spent',	'campaign type': 'campaign_type',	'budget($)': 'budget', 	'duration(days)': 'duration_days'
                            }, inplace=True)
cleaned_df1.head()
```

```python
cleaned_df1.to_csv('cleaned_wholeslanl.csv', index=False)
```
